# **Sesion-05: Distribuciones muestrales y técnicas de evaluación de modelos**

```python
#suroeste_monoxido_de_carbono,noroeste_monoxido_de_carbono,sureste_monoxido_de_carbono,centro_monoxido_de_carbono
co_noreste = CO_mensual['noreste_monoxido_de_carbono']
means = []

for i in range(100_000):
    ej = co_noreste.sample(n=50, replace=True)
    means.append(ej.mean())
    
serie_co_noreste = pd.Series(means)

limite_inferior = serie_co_noreste.quantile(0.025)
limite_superior = serie_co_noreste.quantile(0.975)
print(f'Normalidad de los datos CO_Noreste: {serie_co_noreste.skew()}')
print(f'Curtosis de los datos CO_Noreste: {serie_co_noreste.kurtosis()}')
sns.distplot(serie_co_noreste, kde=False, norm_hist=False)
plt.axvline(limite_inferior)
plt.axvline(limite_superior);
```

**Regresión Linear Múltiple**

```python
from sklearn.model_selection import train_test_split
```

```python
X = CO_mensual[['noreste_monoxido_de_carbono','suroeste_monoxido_de_carbono','noroeste_monoxido_de_carbono','sureste_monoxido_de_carbono']]
y = CO_mensual['centro_monoxido_de_carbono']
```

```python
X_training, X_test, y_training, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_training, y_training)
```

```python
print(f'Precisión: {lr.score(X_test, y_test)}')
```

```python
X = O3_mensual[['noreste_ozono','suroeste_ozono','noroeste_ozono','sureste_ozono']]
y = O3_mensual['centro_ozono']
X_training, X_test, y_training, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_training, y_training)
```

```python
print(f'Precisión: {lr.score(X_test, y_test)}')
```
