# Regresión Lineal

## NO2 Dióxido de nitrógeno

```python
# Paquetes necesarios
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datos/promedios_mensuales/NO2_mensual.csv')
df.head()
```

```python
# Convirtiendo la fecha
import datetime as dt
df['fecha'] = pd.to_datetime(df['fecha'])
df['fecha'] = df['fecha'].map(dt.datetime.toordinal)
```

```python
def obtener_regresion(X, y, zona):
    print(f"\nRegresión Lineal para Dióxido de Nitrógeno (NO2) en {zona}")
    # Dividiendo conjunto de entrenamiento y prueba
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
    
    # Entrenando al modelo
    from sklearn.linear_model import LinearRegression
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    print(f"Score del Modelo: {lr.score(X_train, y_train)}")
    
    # Evaluando modelo de regresión lineal
    y_pred = lr.predict(X_test)
    
    print(" ### Resultados ###")
    print("Conjunto de prueba (test)")
    print(y_test)
    print("Predicción")
    print(y_pred)
    
    # Visualizar los resultados de entrenamiento
    plt.scatter(X_train, y_train, color = "red")
    plt.plot(X_train, lr.predict(X_train), color = "blue")
    plt.title("Contaminación de  Dióxido de Nitrógeno (NO2) en " + zona + " (Entrenamiento)")
    plt.xlabel("Fecha")
    plt.ylabel("Nivel de Contaminación")
    plt.show()
```

```python
X = df.iloc[:,0].values
X = X.reshape(-1,1)
zonas = ["Noreste", "Suroeste", "Noroeste", "Sureste", "Centro"]
for i in range(1,6):
    obtener_regresion(X, df.iloc[:,i].values, zonas[i-1])
```
